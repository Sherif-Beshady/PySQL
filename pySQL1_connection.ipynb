## Connecting to MySQL Database using python
---

_In this notebook, we connect to local MySQL database on a local machine, run a query, restore it in a pandas data frame, add values, and commit these values to the original database and finally, we close the connection._

### Notes:
___

1. You need to install MySQL connector using **"pip install MySQL-connector-python"** in the command line
2. Check the existing databases on your machine using MySQL to make sure the database exists
3. The changes need to be committed using **commit** before you can see the results in MySQL 

```python
import mysql.connector
import pandas as pd
```

```python
con = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'password', #need to change this to your password
    database = "book_shop", #local database
    port = 3306)

print("Hey, I'm connected!")
```

```python
#cursor 
cur = con.cursor()
```

```python
#run simple quary that will fetch table content and put it into dataframe
cur.execute("SELECT * FROM books")
rows = cur.fetchall()
df=pd.DataFrame(rows)
df
```

```python
# Add line to database
cur.execute("INSERT INTO books (title, author_fname, author_lname, released_year, stock_quantity, pages) VALUES('SQL in 48 hours', 'Ryan', 'Stephens',2019, 1000, 475)")

#commit the transaction
con.commit()
```

```python
#close the cursor
cur.close()
```

```python
#close the connection
con.close()
```
